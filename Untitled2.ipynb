```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ---------- 1. MichaelisMenten rate equations ----------

# No inhibitor
def MM(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

# Competitive inhibition: Km is increased by (1 + I/Ki)
def MM_comp(data, Vmax, Km, Ki):
    S, I = data
    return (Vmax * S) / (Km * (1 + I / Ki) + S)

# Non-competitive inhibition: Vmax is decreased by 1 / (1 + I/Ki)
def MM_noncomp(data, Vmax, Km, Ki):
    S, I = data
    return (Vmax / (1 + I / Ki)) * (S / (Km + S))

# ---------- 2. "True" parameters for catalase (arbitrary units) ----------

Vmax_true = 100    # maximum rate
Km_true   = 5      # Michaelis constant
Ki_comp_true    = 3   # competitive inhibition constant
Ki_noncomp_true = 4   # non-competitive inhibition constant

# Substrate concentrations [H2O2] (mM)
S = np.linspace(0.2, 50, 30)

# Inhibitor concentrations (mM)
I_comp    = np.full_like(S, 5.0)   # 5 mM competitive inhibitor
I_noncomp = np.full_like(S, 5.0)   # 5 mM non-competitive inhibitor
I_zero    = np.zeros_like(S)       # no inhibitor

# ---------- 3. Simulate "experimental" data with noise ----------

np.random.seed(0)  # for reproducibility

v_true      = MM(S, Vmax_true, Km_true)
v_comp_true = MM_comp((S, I_comp), Vmax_true, Km_true, Ki_comp_true)
v_non_true  = MM_noncomp((S, I_noncomp), Vmax_true, Km_true, Ki_noncomp_true)

noise_level = 5  # size of random noise

v_no_inhib_data = v_true      + np.random.normal(0, noise_level, size=len(S))
v_comp_data     = v_comp_true + np.random.normal(0, noise_level, size=len(S))
v_non_data      = v_non_true  + np.random.normal(0, noise_level, size=len(S))

# ---------- 4. Fit the data to recover Vmax, Km, Ki ----------

# No inhibition
params_MM, _ = curve_fit(MM, S, v_no_inhib_data, p0=[80, 3])
Vmax_fit, Km_fit = params_MM

# Competitive inhibition
params_comp, _ = curve_fit(MM_comp, (S, I_comp), v_comp_data, p0=[80, 3, 2])
Vmax_comp_fit, Km_comp_fit, Ki_comp_fit = params_comp

# Non-competitive inhibition
params_non, _ = curve_fit(MM_noncomp, (S, I_noncomp), v_non_data, p0=[80, 3, 2])
Vmax_non_fit, Km_non_fit, Ki_non_fit = params_non

print("=== Fitted parameters ===")
print(f"No inhibitor:       Vmax = {Vmax_fit:.2f}, Km = {Km_fit:.2f}")
print(f"Competitive inhib:  Vmax = {Vmax_comp_fit:.2f}, Km = {Km_comp_fit:.2f}, Ki = {Ki_comp_fit:.2f}")
print(f"Non-competitive:    Vmax = {Vmax_non_fit:.2f}, Km = {Km_non_fit:.2f}, Ki = {Ki_non_fit:.2f}")

# ---------- 5. Plot the simulated data and fitted curves ----------

S_fine = np.linspace(0.2, 50, 200)

plt.figure(figsize=(8,6))

# Scatter = noisy data, lines = fitted curves
plt.scatter(S, v_no_inhib_data, color="tab:blue",  label="Data: no inhibitor")
plt.scatter(S, v_comp_data,     color="tab:orange",label="Data: competitive")
plt.scatter(S, v_non_data,      color="tab:green", label="Data: non-competitive")

plt.plot(S_fine, MM(S_fine, Vmax_fit, Km_fit),
         color="tab:blue",  linestyle="-",  label="Fit: no inhibitor")
plt.plot(S_fine, MM_comp((S_fine, np.full_like(S_fine, 5.0)),
                         Vmax_comp_fit, Km_comp_fit, Ki_comp_fit),
         color="tab:orange", linestyle="-", label="Fit: competitive")
plt.plot(S_fine, MM_noncomp((S_fine, np.full_like(S_fine, 5.0)),
                            Vmax_non_fit, Km_non_fit, Ki_non_fit),
         color="tab:green", linestyle="-", label="Fit: non-competitive")

plt.xlabel("[HO] (mM)")
plt.ylabel("Reaction velocity v (a.u.)")
plt.title("Catalase Kinetics with and without Inhibition")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
